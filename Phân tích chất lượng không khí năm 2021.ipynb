```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'historical_air_quality_2021_vn.csv'
dataframe = pd.read_csv(path)
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe['Lượng mưa'] = dataframe[f'Độ ẩm']*10%100
dataframe.head()
```

```python
print(set(dataframe['Lượng mưa']))
```

```python
dataframe['Tháng'] = dataframe[f'Thời gian cập nhật'].str[5:7]
dataframe.head()
```

```python
print(dataframe['Độ ẩm'].dtypes)
print(dataframe['Tốc độ gió'].dtypes)
```

```python
dataframe['Tổng hợp'] = ''
dataframe.head()
```

```python
dataframe['Tổng hợp'] = dataframe['Độ ẩm'] * dataframe['Tốc độ gió']
dataframe.head()
```

```python
a = dataframe.groupby('Tên trạm')['Tốc độ gió'].sum()
a.sort_values()
```

```python
a = dataframe.groupby('Tên trạm')['Lượng mưa'].sum()
a.sort_values()
```

```python
Lượngmưa_values_Têntrạm = dataframe.groupby('Tên trạm').sum()['Lượng mưa']
Lượngmưa_values_Têntrạm.max()
```

```python
Lượngmưa_values_Têntrạm = dataframe.groupby('Tên trạm').sum()['Lượng mưa']
Lượngmưa_values_Têntrạm.min()
```

```python
a = dataframe.groupby('Tháng')['Lượng mưa'].sum()
a.sort_values()
```

```python
a_values_tinh = dataframe.groupby('Tháng').sum()['Độ ẩm']
a_values_tinh.max()
```

```python
a_values_tinh.head()
```

```python
months = range(0,12)
plt.bar(x=months, height=a_values_tinh)
plt.xticks(months, size=10)
plt.xlabel('months')
plt.ylabel('Độ ẩm')
plt.show()
```

```python
a_values_city = dataframe.groupby('Tên trạm').sum()['Lượng mưa']
a_values_city.max()
```

```python
cities = dataframe['Tên trạm'].unique()
plt.bar(x=cities, height=a_values_city)
plt.xticks(cities,rotation=90,size=10)
plt.xlabel('Tên trạm')
plt.ylabel('Độ ẩm')
plt.show()
```
